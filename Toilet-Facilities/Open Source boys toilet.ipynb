# Since we want to analyse boys toilet facilties in schools of India,we get that information from the  data using the code below.

```python
dg=dt.loc[df['State_UT'] == 'All India']
dg
```

# Plot a graph to visualize boys toilet facilities in schools of India.

```python
dg.groupby('year')['Primary_Only', 'U_Primary_Only','Sec_Only',  'HrSec_Only'].sum().plot(kind='bar',figsize = (15, 8), width=0.2)

plt.title('Percentage of boys toilet facilities in India schools',fontsize=15,pad='2.0')
plt.xlabel('Year') 
plt.ylabel('Percentage')
plt.legend(loc='lower right')
plt.show()
```

In 2013-14 school year, the facilities of boys toilets is really low for Higher Secondary Schools but they have a sudden growth from 2014-2015.

Also in 2015-2016, Primary_Only,  U_Primary_Only, Sec_Only and  HrSec_Only  have an increase in boys toilet facilities.

In 2015-2016, the  four schools have an increase in their boys toilet facilities as compared to 2014-2015.

However, this sudden incease in 2015-2016 is realatively small amongst the schools .
